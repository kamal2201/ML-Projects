Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import root_mean_squared_error, accuracy_score
```

Loading the data

```python
wine_data = pd.read_csv('/content/sample_data/winequality-red.csv')
```

```python
wine_data.head()
```

```python
wine_data.isnull().sum()
```

```python
wine_data.describe()
```

```python
wine_data.corr()
```

Splitting the data

```python
X = wine_data.drop(columns='quality', axis=1)
Y = wine_data['quality']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

Training the model

```python
model = RandomForestRegressor()
model.fit(X_train, Y_train)
```

Evaluating the model

```python
training_data_predictions = model.predict(X_train)
training_data_r2_score = model.score(X_train, Y_train)
print('R2 Score:', training_data_r2_score)
```

```python
testing_data_predictions = model.predict(X_test)
testing_data_r2_score = model.score(X_test, Y_test)
print('R2 Score:', testing_data_r2_score)
```

Label Binarizer

```python
Y.head()
```

```python
Y = Y.apply(lambda x: 1 if x >= 7 else 0)
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, Y_train)
```

```python
training_data_predictions = clf.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, training_data_predictions)
print('Accuracy:', training_data_accuracy)
```

```python
testing_data_predictions = clf.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, testing_data_predictions)
print('Accuracy:', testing_data_accuracy)
```
