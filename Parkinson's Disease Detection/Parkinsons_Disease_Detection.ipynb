Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

Loading Dataset

```python
data = pd.read_csv("/content/sample_data/Parkinsons.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.drop(columns = ['name'], axis=1)
```

```python
data.describe()
```

Separating the features

```python
X = data.drop(columns = ['status'], axis=1)
Y = data.status
```

Scaling the values

```python
scaler = StandardScaler()
scaler.fit_transform(X)
```

```python
print(X, Y)
```

Splitting the data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Model Training Evaluation

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
```

```python
train_predictions = logreg.predict(X_train)
training_accuracy = accuracy_score(Y_train, train_predictions)
print("Training Accuracy using Logistic Regression : ", training_accuracy)
```

```python
test_predictions = logreg.predict(X_test)
testing_accuracy = accuracy_score(Y_test, test_predictions)
print("Testing Accuracy using Logistic Regression: ", testing_accuracy)
```

```python
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

```python
train_predictions = model.predict(X_train)
training_accuracy = accuracy_score(Y_train, train_predictions)
print("Training Accuracy using Random Forest Classifier : ", training_accuracy)
```

```python
test_predictions = model.predict(X_test)
testing_accuracy = accuracy_score(Y_test, test_predictions)
print("Testing Accuracy using Random Forest Classifier : ", testing_accuracy)
```

```python
svm = SVC()
svm.fit(X_train, Y_train)
```

```python
train_predictions = svm.predict(X_train)
training_accuracy = accuracy_score(Y_train, train_predictions)
print("Training Accuracy using SVM : ", training_accuracy)
```

```python
test_predictions = svm.predict(X_test)
testing_accuracy = accuracy_score(Y_test, test_predictions)
print("Testing Accuracy using SVM : ", testing_accuracy)
```
