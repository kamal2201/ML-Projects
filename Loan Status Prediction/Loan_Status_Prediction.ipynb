Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

Downloading Dataset

```python
loan_data = pd.read_csv('/content/sample_data/Loan_Data.csv')
```

Preprocessing Dataset

```python
loan_data.head()
```

```python
loan_data.isnull().sum()
```

```python
loan_data = loan_data.dropna()
```

```python
loan_data.shape
```

```python
encoder = LabelEncoder()
loan_data['Gender'] = encoder.fit_transform(loan_data['Gender'])
loan_data['Married'] = encoder.fit_transform(loan_data['Married'])
loan_data['Dependents'] = encoder.fit_transform(loan_data['Dependents'])
loan_data['Education'] = encoder.fit_transform(loan_data['Education'])
loan_data['Self_Employed'] = encoder.fit_transform(loan_data['Self_Employed'])
loan_data['Property_Area'] = encoder.fit_transform(loan_data['Property_Area'])
loan_data['Loan_Status'] = encoder.fit_transform(loan_data['Loan_Status'])
```

```python
loan_data.info()
```

Splitting the data

```python
X = loan_data.drop(columns=['Loan_ID', 'Loan_Status'], axis=1)
Y = loan_data['Loan_Status']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

Model training

```python
model = SVC(kernel='linear')
model.fit(X_train, Y_train)

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
```

Model Evaluation

```python
train_data_predictions = logreg.predict(X_train)
train_data_accuracy = accuracy_score(Y_train, train_data_predictions)
print("Training Accuracy using Logistic Regression : ", train_data_accuracy)
```

```python
test_data_predictions = logreg.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, test_data_predictions)
print("Testing Accuracy using Logistic Regression : ", test_data_accuracy)
```

```python
train_data_predictions = model.predict(X_train)
train_data_accuracy = accuracy_score(Y_train, train_data_predictions)
print("Training Accuracy using SVM : ", train_data_accuracy)
```

```python
test_data_predictions = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, test_data_predictions)
print("Testing Accuracy using SVM : ", test_data_accuracy)
```
